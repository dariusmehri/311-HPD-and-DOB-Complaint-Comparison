```python
import pandas as pd
import numpy as np

path = "C:\\Users\\dmehri\\Documents\\DATA\\311 Analysis\\"

agencylist = ["HPD", "DOB"]

df0 = pd.read_csv(path + "311_Service_Requests_2017.csv", low_memory=False)
df0 = df0[df0['Agency'].isin(agencylist)].reset_index(drop=True)

df1 = pd.read_csv(path + "311_Service_Requests_2018.csv", low_memory=False)
df1 = df1[df1['Agency'].isin(agencylist)].reset_index(drop=True)

df2 = pd.read_csv(path + "311_Service_Requests_2019.csv", low_memory=False)
df2 = df2[df2['Agency'].isin(agencylist)].reset_index(drop=True)

df3 = pd.read_csv(path + "311_Service_Requests_2020.csv", low_memory=False)
df3 = df3[df3['Agency'].isin(agencylist)].reset_index(drop=True)

df4 = pd.read_csv(path + "311_Service_Requests_2021.csv", low_memory=False)
df4 = df4[df4['Agency'].isin(agencylist)].reset_index(drop=True)

frames = [df0, df1, df2, df3, df4]

df = pd.concat(frames)
df = df.reset_index(drop=True)
```

```python
subset = ['Unique Key', 'Created Date', 'Closed Date', 'Agency', 'Agency Name',
       'Complaint Type', 'Descriptor', 'Location Type', 'Incident Zip',
       'Incident Address', 'Street Name', 'City', 'Community Board','Latitude', 'Longitude']

df = df[subset]
```

```python
df.to_csv(path + "311_Service_Requests_HPD_DOB_2017-21.csv")
```

```python
df = pd.read_csv(path + "311_Service_Requests_HPD_DOB_2017-21.csv", low_memory=False)
```

```python
pd.set_option('chained_assignment', None)

df["Complaint Type"] = df["Complaint Type"].map(str.upper)
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Created Date'] = df['Created Date'].apply(lambda x: x.date())

df['Closed Date'] = pd.to_datetime(df['Closed Date'])
df['Closed Date'] = df['Closed Date'].apply(lambda x: x.date())

df["Year"] = pd.DatetimeIndex(df['Created Date']).year
```

Complaint Types

```python
dfg = df[["Agency","Year", "Complaint Type"]]
dfg["Count"] = 1
dfg = dfg.groupby(["Agency","Year", "Complaint Type"]).sum()
dfg = dfg.add_suffix('').reset_index()

dfg = dfg.sort_values(by = 'Count', ascending=False).reset_index(drop=True)
```

```python
dfg.to_csv(path + "311 HPD DOB Complaint Type.csv", index=False)
```

Service Levels

```python
#df["Business Days to Close"] = np.busday_count( df['Closed Date'].values.astype('datetime64[D]'), df['Created Date'].values.astype('datetime64[D]'))
```

```python
df['Days to Close'] = (df["Closed Date"] - df["Created Date"]).dt.days

#df["Business Days to Close"] = np.busday_count( df['Closed Date'].values.astype('datetime64[D]'), df['Created Date'].values.astype('datetime64[D]'))


dfg2 = df[["Agency","Year", "Complaint Type", "Days to Close"]]
dfg2 = dfg2.groupby(["Agency","Year", "Complaint Type"]).mean()
dfg2 = dfg2.add_suffix('').reset_index()


```

```python
dfg2.to_csv(path + "311 HPD DOB Complaint Type Service Levels.csv", index=False)
```

Service Levels at Community Board

```python
df["Community Board"] = df["Community Board"].astype(str)

cblist = df["Community Board"].tolist()
#a = cblist[0]
newcblist = []
for i in range(0, len(cblist)):
    a = cblist[i]
    #print (a)
    if "MANHATTAN" in a:
        a = a.split(" ")
        a = "1" + str(a[0])
        #print (a)
        newcblist.append(a)
    elif "BRONX" in a:
        a = a.split(" ")
        a = "2" + str(a[0])
        #print (a)
        newcblist.append(a)
    elif "BROOKLYN" in a:
        a = a.split(" ")
        a = "3" + str(a[0])
        #print (a)
        newcblist.append(a)
    elif "QUEENS" in a:
        a = a.split(" ")
        a = "4" + str(a[0])
        #print (a)
        newcblist.append(a)
    elif "STATEN ISLAND" in a:
        a = a.split(" ")
        a = "5" + str(a[0])
        #print (a)
        newcblist.append(a)
    else:
        newcblist.append(a)
        
        
    
  
print (len(df))  
print(len(newcblist) )

dfcb = pd.DataFrame(newcblist)
dfcb = dfcb.rename(columns={0: 'CB'})

print ("len of dfcb")
print (len(dfcb))


df = pd.concat([df, dfcb], axis=1)

print ("len of df")
print (len(df))
```

Drop Rows with Unspecified CB

```python
print (len(df))
df2 = df[df.CB.str.contains("Unspecified") == False]
print (len(df2))
```

```python
dfc = df2[["Agency","CB","Days to Close"]]
dfc = dfc.groupby(["Agency","CB"]).mean()
dfc = dfc.add_suffix('').reset_index()
```
